```python
import pandas as pd
import os
import shutil
```

```python
data = pd.read_csv("lfw-names.txt", sep='\t', header=None)
data.columns = ['name', 'count']
```

```python
data
```

```python
data2 = data[data['count']>3]
```

```python
data2.to_csv('lfw-names.txt', sep='\t', index=False)
```

```python
file = os.path.join(os.getcwd(), os.listdir(os.getcwd())[1])
file
```

```python
lfw_dir = [x[0] for x in os.walk(file)]
```

```python
lfw_dir[1:]
```

```python
for d in lfw_dir[1:]:
    name = d.split('\\')[5]
    if name not in data2['name'].tolist():
        shutil.rmtree(d)
        print('remove', name)
    else:
        print('keep', name)

```
